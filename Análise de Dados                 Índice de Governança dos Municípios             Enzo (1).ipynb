```python
# Criando um dataframe com base no arquivo excel
# Precisamos pular as 2 primeiras linhas (skiprows=2) e as
# ultimas 16 linhas (skipfooter=16)
df = pd.read_excel("dados_IGM-CFA_2016.xlsx", skiprows=1, skipfooter=16)

# Exibindo as 5 primeiras linhas
df.head()
```

```python
# Olhando a quantidade de dados (L,C)
df.shape
```

```python
# Olhando o tipo e quantidade
df.info()
```

```python
#Qual a maior população
df['População 2016 (IBGE)'].max()
```

```python
#Qual o município com a maior população
df[df['População 2016 (IBGE)'] == df['População 2016 (IBGE)'].max()]
```

```python
#Qual o município com a menor população
df[df['População 2016 (IBGE)'] == df['População 2016 (IBGE)'].min()]
```

```python
#Qual o município com a maior densidade demográfica
df[df['Densidade demográfica 2010 (hab/km²) (IBGE Cidades)'] == 
   df['Densidade demográfica 2010 (hab/km²) (IBGE Cidades)'].max()]
```

```python
#Qual o município com a menor densidade demográfica
df[df['Densidade demográfica 2010 (hab/km²) (IBGE Cidades)'] == 
   df['Densidade demográfica 2010 (hab/km²) (IBGE Cidades)'].min()]
```

```python
#Máxima de População por Estado
ax = df.groupby('Estado').max()['População 2016 (IBGE)'].plot.bar(figsize = (10, 5))
ax.set_ylabel('População 2016 (IBGE)')
```

```python
#Qualidade Habitacional por Estado
ax = df.groupby('Estado').max()['Qualidade habitacional'].plot.bar(figsize = (10, 5))
ax.set_ylabel('Qualidade habitacional')
```

```python
#Qualidade Habitacional por Estado
ax = df.groupby('Estado').max()['Vulnerabilidade'].plot.bar(figsize = (10, 5))
ax.set_ylabel('Vulnerabilidade')
```

```python
#Mortalidade Infantil
ax = df.groupby('Estado').max()['Mortalidade infantil'].plot.bar(figsize = (10, 5))
ax.set_ylabel('Mortalidade infantil')
```

```python
#Qual o município com a menor Mortalidade Infantil
df[df['Mortalidade infantil'] == df['Mortalidade infantil'].min()]
```

```python
#Qual o município com a maior Mortalidade Infantil
df[df['Mortalidade infantil'] == df['Mortalidade infantil'].max()]
```
